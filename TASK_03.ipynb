```python
high_corr_col=['emp.var.rate','euribor3m','nr.employed']
```

```python
df1=data.copy()
df1.columns
```

```python
df1.drop(high_corr_col,inplace=True,axis=1)
df1.columns
```

```python
df1.shape
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df_encoded=df1.apply(lb.fit_transform)
df_encoded
```

```python
df_encoded['deposit'].value_counts()
```

```python
x=df_encoded.drop('deposit',axis=1)
y=df_encoded['deposit']
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size= 0.25,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix,classification_report, accuracy_score
def eval_model(y_test,y_pred):
    acc=accuracy_score(y_test,y_pred) 
    print('Accuracy score',acc)
    ca =confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',ca)
    print('Classification Reportin', classification_report(y_test,y_pred))
def ascore(model):
    train_score= model.score(x_train,y_train)
    test_score= model.score(x_test,y_test)
    print("Training score", train_score)
    print("Testing Score", test_score)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion ='gini',max_depth =5,min_samples_split=10)
dt.fit(x_train,y_train)
```

```python
ascore(dt)
```

```python
ypred_dt=dt.predict(x_test)
print(ypred_dt)
```

```python
eval_model(y_test,ypred_dt)
```

```python
from sklearn.tree import plot_tree
cn=['no','yes']
fn=x_train.columns
print(fn)
print(cn)
```

```python
plt.figure(figsize=(40,20))
plot_tree(dt,class_names=cn,filled=True)
plt.show()
```

```python
dt1=DecisionTreeClassifier(criterion ='entropy',max_depth =4,min_samples_split=15)
dt1.fit(x_train,y_train)
```

```python
ascore(dt1)
```

```python
ypred_dt1=dt1.predict(x_test)
eval_model(y_test,ypred_dt1)
```

```python
plt.figure(figsize=(40,20))
plot_tree(dt1,class_names=cn,filled=True)
plt.show()
```
